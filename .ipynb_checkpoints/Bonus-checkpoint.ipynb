```python
import pandas as pd
import matplotlib.pyplot as plt

from sqlalchemy import create_engine
```

```python

engine = create_engine('postgresql://postgres:@localhost:5432/HW_draft')
connection = engine.connect()
```

```python
emp=pd.read_sql("SELECT * FROM employees ", connection)
```

```python
salary=pd.read_sql("SELECT * FROM salaries", connection)
```

```python
emp.head()
```

```python
salary.head()
```

```python
emp_df=pd.merge(emp,salary,how='outer',on='emp_no')
emp_df.head()
```

```python
# Need to clean up the histogram
hist=emp_df["salary"].hist(bins=3)
```

```python
title=pd.read_sql("SELECT * FROM titles", connection)
```

```python
title.head()
```

```python
emp_df.head()
```

```python
emply=emp_df.rename(columns={"emp_title_id":"title_id"})
```

```python
emply.head()
```

```python
empl_title_df=pd.merge(emply,title,how='outer',on='title_id')
empl_title_df.head()
```

```python
title_df=empl_title_df.groupby('title')
```

```python
title_df.head()
```

```python
avg_salary=round(title_df["salary"].mean(),2)
avg_salary
```

```python
avg_salary_df=pd.DataFrame({"Average Salary":avg_salary})
avg_salary_df
```

```python
avg_salary.plot(kind="bar")
plt.title("Average pay by title")
plt.ylabel("Salary")
plt.show()
```
